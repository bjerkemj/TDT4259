```python
import pandas as pd
from autots import AutoTS
```

```python
# load datta
data = pd.read_csv('data_NO2_boosted.csv')
data = data.iloc[:-2]
data.tail()
```

```python
# Ensure the date column is in datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the date as the index
data.set_index('date', inplace=True)

train_data = data.iloc[:-208*24]
test_data = data.iloc[-208*24:]

## Initialize AutoTS with specified parameters
model = AutoTS(
    forecast_length=24,          # Forecast horizon
    frequency='H',               # Frequency of the data
    ensemble='simple',           # Use ensemble of models for better accuracy
    model_list='fast',           # Use fast model list for quicker results
    max_generations=5            # Number of genetic algorithm generations to evolve
)

# Fit the model on the training data
model.fit(train_data[['spot_price']])

# Make predictions on the test data
predictions = model.predict()
forecast = predictions['predicted_dataframe']

# Display the forecast and test set for comparison
print(forecast)
print(test_data[['spot_price']])
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Ensure predictions and test_data are aligned
predicted_values = forecast['spot_price'].values  # Extract predicted values from the forecast
actual_values = test_data['spot_price'].values     # Actual values from the test data

# Calculate MAE
mae = mean_absolute_error(actual_values, predicted_values)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(actual_values, predicted_values))

# Display the results
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
